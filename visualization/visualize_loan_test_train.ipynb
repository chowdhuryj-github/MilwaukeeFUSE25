## <span style="color: #04A4E2"> Data Visualization: Scatter Plot, Joint Plot or Kernel Density Estimation Plot </span>
To start, we first focus on numerical features that we're interested in and determining the relationships between those features. Here are the features that we're interested in looking at:

 - The relationship between `ApplicantIncome` & `Loan_Amount_Term` using a Scatter Plot

### <span style="color: #04A4E2"> Defining a Method for Scatter Plot </span>
For ease and reducing code repitition, the cell below will define a method for creating scatter plots, include the title, labels for x and y axis, as well as any other related plotting information.

```python
# method for plotting out a scatter plot
def plot_scatter_plot(dataframe, feature_one, feature_two):
    plt.scatter(dataframe[feature_one], dataframe[feature_two])
    plt.xlabel(feature_one)
    plt.ylabel(feature_two)
    plt.title(f"{feature_one} vs {feature_two} Scatter Plot")
    plt.show()
```

### <span style="color: #04A4E2"> Creating the Scatter Plots </span>
Now, we work on creating thr scatter plots using the features we're interested in. We achieve this by calling our previous method.

```python
# calling the method to plot out the scatter plots
plot_scatter_plot(clean_train_df, "ApplicantIncome", "LoanAmount")
plot_scatter_plot(clean_train_df, "Loan_Amount_Term", "ApplicantIncome")
plot_scatter_plot(clean_train_df, "LoanAmount", "CoapplicantIncome")
plot_scatter_plot(clean_train_df, "Loan_Amount_Term", "CoapplicantIncome")
```

### <span style="color: #04A4E2"> Interpretations from the Scatter Plot </span>
After creating the visualizations, these are some of the interpretations that can be derived from the scatter plots:
 - `ApplicantIncome` vs `LoanAmount`: there is a positive linear relationship between these two features
 - `Loan_Amount_Term` vs `ApplicantIncome`: there is no linear relationship between these two featurees
 - `LoanAmount` vs `CoapplicantIncome`: there is a positive linear relationship between these two features
 - `Loan_Amount_Term` vs `CoapplicantIncome`: there is no linear relationship between these two features

## <span style="color: #F4AB04"> Data Visualization: Boxplots & Violin Plots </span>
Next, we focus on looking at categorical and numerical features, as we'd like to visualize the features of interest to us, so that we can determine the relationships between these feature. The methods of visualization we'd like to use are box plots and violin plots. Here are the features we're interested in:
- The relationship between `ApplicantIncome` and `Loan_Status` using a box plot
- The relationship between `LoanAmount` and `Loan_Status` using a box plot
- The relationship between `Loan_Amount_Term` and `Loan_Status` using a violin plot
- The relationship between `CoapplicantIncome` and `Loan_Status` using a violin plot

### <span style="color: #F4AB04"> Defining a Method for plotting Box Plots or Violin Plots </span>
We start off by first defining a method for plotting out box plots or violin plots. This helps us to avoid writing repeat code and simply call the method with the features we're interested in as parameters.

```python
# method for plotting out a box or violin plot
def plot_violin_box_plot(dataframe, categorical_feature, numerical_feature, type):

    if type == "box":
        sns.boxplot(x=categorical_feature, y=numerical_feature, data=dataframe)
        plt.title(f"{type} plot for {categorical_feature} vs {numerical_feature}")
        plt.xlabel(categorical_feature)
        plt.ylabel(numerical_feature)
        plt.show()

    elif type == "violin":
        sns.violinplot(x=categorical_feature, y=numerical_feature, data=dataframe)
        plt.title(f"{type} plot for {categorical_feature} vs {numerical_feature}")
        plt.xlabel(categorical_feature)
        plt.ylabel(numerical_feature)
        plt.show()
```

### <span style="color: #F4AB04"> Creating the Box and Violin Plots </span>
Now, we focus on creating the box and violin plots, Looking at these data visualizations will allow us to analyze them and determine how the relationships between these features look like as shown below.

```python
plot_violin_box_plot(clean_train_df, "Loan_Status", "ApplicantIncome", "box")
plot_violin_box_plot(clean_train_df, "Loan_Status", "LoanAmount", "box")
plot_violin_box_plot(clean_train_df, "Loan_Status", "Loan_Amount_Term", "violin")
plot_violin_box_plot(clean_train_df, "Loan_Status", "CoapplicantIncome", "violin")
```

### <span style="color: #F4AB04"> Interpretations from the Box & Violin Plots </span>
Now that we've created our box and violin plots, we'd like to analyze these plots which will allow us to gain a better understanding:
- `Loan_Status` vs `ApplicantIncome`: we can see that the median is very similar between these two box plots
- `Loan_Status` vs `LoanAmount`: we can see that there may be a slight difference between the medians of these two box plots
- `Loan_Status` vs `Loan_Amount_Term`: we can see that the median is very similar between these two violin plots
- `Loan_Status` vs `CoapplicantIncome`: we can see that there may be a slight difference between these two violin plots
